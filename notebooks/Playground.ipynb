# Interpolation

```python
from deepsvg.svglib.svg import SVG

from deepsvg import utils
from deepsvg.difflib.tensor import SVGTensor
from deepsvg.svglib.utils import to_gif
from deepsvg.svglib.geom import Bbox
from deepsvg.svgtensor_dataset import SVGTensorDataset, load_dataset
from deepsvg.utils.utils import batchify, linear

import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
pretrained_path = "./pretrained/hierarchical_ordered.pth.tar"
from configs.deepsvg.hierarchical_ordered import Config

cfg = Config()
cfg.print_params()
model = cfg.make_model().to(device)
utils.load_model(pretrained_path, model, device)
model.eval()
```

```python
dataset, _ = load_dataset(cfg)
```

```python
def easein_easeout(t):
    return t*t / (2. * (t*t - t) + 1.)

def interpolate(z1, z2, n=25, filename=None, ease=True, do_display=True):
    alphas = torch.linspace(0., 1., n)
    if ease:
        alphas = easein_easeout(alphas)
    z_list = [(1-a) * z1 + a * z2 for a in alphas]
    
    img_list = [decode(z, do_display=False, return_png=True) for z in z_list]
    to_gif(img_list + img_list[::-1], file_path=filename, frame_duration=1/12)
```

```python
def encode_icon(idx):
    data = dataset.get(id=idx, random_aug=False)
    model_args = batchify((data[key] for key in cfg.model_args), device)
    with torch.no_grad():
        z = model(*model_args, encode_mode=True)
    return z

def decode(z, do_display=True, return_svg=False, return_png=False):
    commands_y, args_y = model.greedy_sample(z=z)
    tensor_pred = SVGTensor.from_cmd_args(commands_y[0].cpu(), args_y[0].cpu())
    svg_path_sample = SVG.from_tensor(
        tensor_pred.data, viewbox=Bbox(256), allow_empty=True
    ).normalize().split_paths().set_color("random")
    
    if return_svg:
        return svg_path_sample
    
    return svg_path_sample.draw(do_display=do_display, return_png=return_png)
    
```

```python
def interpolate_icons(idx1=None, idx2=None, n=25, *args, **kwargs):
    z1, z2 = encode_icon(idx1), encode_icon(idx2)
    interpolate(z1, z2, n=n, *args, **kwargs)
```

```python
from deepsvg.svglib.svg import SVG

from deepsvg import utils
from deepsvg.difflib.tensor import SVGTensor
from deepsvg.svglib.utils import to_gif
from deepsvg.svglib.geom import Bbox
from deepsvg.svgtensor_dataset import SVGTensorDataset, load_dataset
from deepsvg.utils.utils import batchify, linear

import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
pretrained_path = "./pretrained/hierarchical_ordered.pth.tar"
from configs.deepsvg.hierarchical_ordered import Config

cfg = Config()
cfg.print_params()
model = cfg.make_model().to(device)
utils.load_model(pretrained_path, model, device)
model.eval()
```

```python
dataset, _ = load_dataset(cfg)
for k,t in dataset.random_icon().items():
    print(f"{k+':':15s} {t.shape}")
```

```python
ma_model_args = ["args", "commands", "args_grouped", "commands_grouped", "args_rel", "filling", "label"]

random_idx = 72
svg_tensor = dataset.get(random_idx, model_args=ma_model_args)
tensor_list = [svg_tensor[k] for k in ma_model_args]

for k, t in zip(ma_model_args, tensor_list):
    print(f"{k+':':15s} {t.shape}")

print("\n   BATCHIFY:")
for k, t in zip(ma_model_args, batchify(tensor_list, device)):
    print(f"{k+':':15s} {t.shape}")
```

```python
#~~~ Starting with one datapoint ~~~#

N_POINTS = 10

# prepare data
data = dataset.get(id=random_idx, random_aug=False, model_args=["commands", "args", "tensor_grouped"])
print(f"data.keys()={data.keys()}")
print(f"data['tensor_grouped']={data['tensor_grouped']}")
print(f"data['tensor_grouped'][0]={data['tensor_grouped'][0]}")
print(f"cfg.model_args={cfg.model_args}")


# target points
tensor_target = data["tensor_grouped"][0].copy().drop_sos().unpad()
points_target = tensor_target.sample_points(n=N_POINTS)

# greedy sample prediction
model_args = batchify((data[key] for key in cfg.model_args), device)
with torch.no_grad():
    z = model(*model_args, encode_mode=True)
print(f"z.shape={z.shape}")

commands_y, args_y = model.greedy_sample(z=z)


# prediction points
tensor_pred = SVGTensor.from_cmd_args(commands_y[0], args_y[0])
print(f"commands_y.shape={commands_y.shape}")
print(f"args_y.shape={args_y.shape}")

points_pred = tensor_pred.sample_points(n=N_POINTS)


# reconstruction loss
print("TARGET:")
SVG.from_tensor(tensor_target.data, viewbox=Bbox(256)).normalize().split_paths().set_color("random").draw()
print("PREDICTION:")
SVG.from_tensor(tensor_pred.data, viewbox=Bbox(256)).normalize().split_paths().set_color("random").draw()

print(f"Reconstruction losses")
print(f"Chamfer loss:               {chamfer_loss(points_pred, points_target)}")
print(f"Earth movers distance loss: {svg_emd_loss(points_pred, points_target)}")
```

```python
from torch.utils.data import DataLoader

#~~~ Reconstruction loss for a batch ~~~#

# print(data.keys())
# tensor_withoug_start_pos = torch.cat([
#     data["commands"].unsqueeze(-1),
#     data["args"]
# ], axis=-1).tolist()[:15]
# tensor = torch.cat([
#     tensor_withoug_start_pos[SVGTensor.IndexArgs.END_POS]
# ])
# svgtensor = SVGTensor.from_cmd_args(data["commands_grouped"], data["args_grouped"])
# svgtensor.drop_sos().unpad()
# svgtensor.data.tolist()[:15]

# commands = data["commands_grouped"]
# args = data["args_grouped"]
# svgtensor = SVGTensor(
#     commands,
#     args[..., SVGTensor.IndexArgs.RADIUS],
#     args[..., SVGTensor.IndexArgs.X_AXIS_ROT],
#     args[..., SVGTensor.IndexArgs.LARGE_ARC_FLG],
#     args[..., SVGTensor.IndexArgs.SWEEP_FLG],
#     args[..., SVGTensor.IndexArgs.CONTROL1],
#     args[..., SVGTensor.IndexArgs.CONTROL2],
#     args[..., SVGTensor.IndexArgs.END_POS]
# )
# svgtensor.data

def batch_reconstruction_loss_for_svg_sampled_points(model, cfg, batch, loss_fn_dict = {"loss_re": chamfer_loss, "loss_emd":svg_emd_loss}, show_logs=False):
    commands = batch["commands_grouped"]
    args = batch["args_grouped"]
    losses = {k:[] for k in loss_fn_dict.keys()}
    for i, (c, a) in enumerate(zip(commands, args)):
        # target points
        tensor_target = SVGTensor.from_cmd_args(c[0], a[0]).copy().unpad().drop_sos() # TODO Do I need to copy it?
        points_target = tensor_target.sample_points(n=N_POINTS)

        # greedy sample prediction
        model_args = batchify((batch[key][i] for key in cfg.model_args), device)
        with torch.no_grad():
            z = model(*model_args, encode_mode=True)

        commands_y, args_y = model.greedy_sample(z=z)


        # prediction points
        tensor_pred = SVGTensor.from_cmd_args(commands_y[0], args_y[0])
        points_pred = tensor_pred.sample_points(n=N_POINTS)


        # reconstruction loss
        for k, loss_fn in loss_fn_dict.items():
            losses[k].append(loss_fn(points_pred, points_target))

        if show_logs:
            print(f"Reconstruction losses")
            for k in loss_fn_dict.keys():
                print(f"LOSS {i:04} {k}:\t{losses[k][-1]}")

            if i < 10:
                print("TARGET:")
                SVG.from_tensor(tensor_target.data, viewbox=Bbox(256)).normalize().split_paths().set_color("random").draw()
                print("PREDICTION:")
                SVG.from_tensor(tensor_pred.data, viewbox=Bbox(256)).normalize().split_paths().set_color("random").draw()

    return losses
```

```python
N_POINTS = 10

# reset config
pretrained_path = "./pretrained/hierarchical_ordered.pth.tar"
cfg = Config()
cfg_modified = Config()
cfg_modified.model_args += ["commands_grouped", "args_grouped"]

# prepare data
dataset, _ = load_dataset(cfg_modified)
dataloader = DataLoader(
    dataset,
    batch_size=cfg.batch_size,
    shuffle=True,
    drop_last=True,
    num_workers=cfg.loader_num_workers,
    collate_fn=cfg.collate_fn,
)
data = next(iter(dataloader))
print(data.keys())
print("COMPUTING THE LOSSES")
loss_dict = batch_reconstruction_loss_for_svg_sampled_points(model, cfg, data, show_logs=True)
```

```python
from deepsvg.utils import Stats
import tqdm
import pandas as pd

#~~~ The whole dataset ~~~#
split="valid"
step=1
epoch=1
stats = Stats(
    num_steps=cfg.num_steps,
    num_epochs=cfg.num_epochs,
    steps_per_epoch=len(dataloader),
    stats_to_print=cfg.stats_to_print
)

cfg_modified.train_ratio=0.999
_, dataset = load_dataset(cfg_modified)
dataloader = DataLoader(
    dataset,
    batch_size=cfg.batch_size,
    shuffle=True,
    drop_last=True,
    num_workers=cfg.loader_num_workers,
    collate_fn=cfg.collate_fn,
)

def reconstruction_loss_for_svg_sampled_points(model, cfg, dataloader, loss_fn_dict = {"loss_re": chamfer_loss, "loss_emd":svg_emd_loss}, show_logs=False):
    model.eval()
    with torch.no_grad():
        for data in tqdm.tqdm(dataloader):
            losses_dict = batch_reconstruction_loss_for_svg_sampled_points(model, cfg, data, loss_fn_dict, show_logs)
            for k, losses in losses_dict.items():
                df = pd.DataFrame(losses).describe().transpose()
                loss_dict = {k+"/"+id:float(df[id]) for id in ["mean", "std", "min", "25%", "50%", "75%", "max"]}
                stats.update_stats_to_print(split, loss_dict.keys())
                stats.update(split, step, epoch, {**loss_dict})
            print(stats.get_summary(split))

reconstruction_loss_for_svg_sampled_points(model, cfg, dataloader)
```
