```python


from deepsvg import utils
from deepsvg.svgtensor_dataset import load_dataset

import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
# prepare the data
from configs.deepsvg.hierarchical_ordered import Config
from torch.utils.data import DataLoader

cfg = Config()
cfg_modified = Config()
cfg_modified.model_args += ["commands_grouped", "args_grouped"]
dataset, _ = load_dataset(cfg_modified)
dataloader = DataLoader(
    dataset,
    batch_size=24,  #cfg.batch_size,
    shuffle=True,
    drop_last=True,
    num_workers=cfg.loader_num_workers,
    collate_fn=cfg.collate_fn,
)
```

```python
pretrained_path = "./pretrained/hierarchical_ordered.pth.tar"
from configs.deepsvg.hierarchical_ordered import Config

cfg = Config()
cfg.print_params()
model = cfg.make_model().to(device)
utils.load_model(pretrained_path, model, device)
model.eval()
print("Model loaded")
```

```python
from deepsvg.train import _batch_reconstruction_loss_for_svg_sampled_points
from deepsvg.difflib.loss import svg_emd_loss
from deepsvg.difflib.loss import chamfer_loss

N_POINTS = 20
cfg.n_recon_points = N_POINTS
cfg.loss_recon_fn_dict = {"loss_re": chamfer_loss, "loss_emd": svg_emd_loss}

batch = next(iter(dataloader))
print("COMPUTING THE LOSSES")
result = _batch_reconstruction_loss_for_svg_sampled_points(model, cfg, batch, show_logs=True)
```

```python
import numpy as np
loss_re = np.array(result["loss_re"])
loss_re.mean()
```
